```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn
import sklearn
import sklearn.ensemble
import sklearn.compose
from sklearn.preprocessing import *
from sklearn.feature_selection import VarianceThreshold

data = pd.read_csv("Automobile\\Automobile_data.csv")
data
```

```python
data.isnull().sum()
```

```python
label_encoder = LabelEncoder()

data = data.replace("?", None)

data['make'] = label_encoder.fit_transform(data['make'])
data['fuel-type'] = label_encoder.fit_transform(data['fuel-type'])
data['aspiration'] = label_encoder.fit_transform(data['aspiration'])
data['num-of-doors'] = label_encoder.fit_transform(data['num-of-doors'])
data['body-style'] = label_encoder.fit_transform(data['body-style'])
data['drive-wheels'] = label_encoder.fit_transform(data['drive-wheels'])
data['engine-location'] = label_encoder.fit_transform(data['engine-location'])
data['engine-type'] = label_encoder.fit_transform(data['engine-type'])
data['num-of-cylinders'] = label_encoder.fit_transform(data['num-of-cylinders'])
data['fuel-system'] = label_encoder.fit_transform(data['fuel-system'])

data = data.fillna(value=0)
```

```python
data.isnull().sum()
```

```python
data['price'] = data['price'].apply(lambda x: float(x))
display(data.dtypes)
data
```

```python
seaborn.heatmap(data.corr())
```

```python
c_d = data[data.price < data.price.quantile(0.9)]
c_d
```

```python

col_arr = data.columns

selection = VarianceThreshold(threshold=(0.5))
data_selected = selection.fit_transform(c_d.drop(["price"], axis=1), c_d["price"])
features_selected = selection.get_feature_names_out(col_arr[:len(col_arr) - 1])
data_selected = pd.DataFrame(data_selected, columns=features_selected)

features_deleted = np.array([], dtype=str)

for feature in col_arr[:len(col_arr) - 1]:
    if feature not in features_selected:
        features_deleted = np.append(features_deleted, feature)

display(pd.DataFrame(features_selected))
display(pd.DataFrame(features_deleted))
display(data_selected)
```

```python
tree_classifier = sklearn.ensemble.RandomForestRegressor(n_estimators=50)
tree_classifier.fit(c_d.drop(["price"], axis=1), c_d["price"])
model = sklearn.feature_selection.SelectFromModel(tree_classifier, prefit=True)
```

```python
feature_importances = tree_classifier.feature_importances_
argsort = np.argsort(feature_importances)
plt.yticks(range(len(argsort)), np.array(c_d.drop(["price"], axis=1).columns)[argsort])
plt.barh(range(len(argsort)), feature_importances[argsort])
plt.show()
```
