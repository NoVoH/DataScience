```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_excel("dds_ch2_rollingsales\\rollingsales_bronx.xls", header = 4)

data
```

```python
data.drop("EASE-MENT", inplace = True, axis = 1)

data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data['ZIP CODE'] = data['ZIP CODE'].fillna(0)
data['YEAR BUILT'] = data['YEAR BUILT'].fillna(0)
data['TAX CLASS AT PRESENT'] = data['TAX CLASS AT PRESENT'].fillna(0)
data['BUILDING CLASS CATEGORY'] = data['BUILDING CLASS CATEGORY'].fillna(0)
data['NEIGHBORHOOD'] = data['NEIGHBORHOOD'].fillna(0)
data 
```

```python
data = data[data['SALE\nPRICE'] != 0]
data
```

```python
X = data["LAND SQUARE FEET"].values.reshape(-1, 1)
Y = data["GROSS SQUARE FEET"].values.reshape(-1, 1)
linear = LinearRegression()
linear.fit(X, Y)
Y_pred = linear.predict(X)

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.xlabel("LAND SQUARE FEET")
plt.ylabel("GROSS SQUARE FEET")
plt.show()

data.drop("GROSS SQUARE FEET", inplace = True, axis = 1)
```

```python
grouped_data = data.groupby("NEIGHBORHOOD")
grouped_data = grouped_data.mean()
grouped_data["SALE\nPRICE"].plot.bar(figsize=(15, 5), ylabel="Sale price")
```

```python
def getSales(start_date, end_date): 
    bounds = (data["SALE DATE"] >= start_date) & (data["SALE DATE"] < end_date)
    sales = data.loc[bounds]
    return sales

start_date = "2012-10-01"
end_date = "2013-01-01"

quarter_sales1 = getSales(start_date, end_date).groupby("NEIGHBORHOOD")
quarter_sales1["BOROUGH"].count().plot.bar(ylabel="Number of houses sold from " + start_date + " - " + end_date, figsize=(15, 5))
```

```python
def year_division(year):
    if year == 0:
        return "Date unknown"
    elif year <= 1800:
        return "<=1800"
    elif year <= 1850:
        return "1801-1850"
    elif year <= 1900:
        return "1851-1900"
    elif year <= 1950:
        return "1901-1950"
    elif year <= 2000:
        return "1951-2000"
    else:
        return "2000>"

data["YEAR BUILT GROUP"] = data["YEAR BUILT"].apply(lambda x: year_division(x))
data
```

```python
data.groupby("YEAR BUILT GROUP").count().plot.bar(y="BOROUGH", xlabel="Time when wasbuilt", ylabel="Number of buildings", figsize=(5,15))
```

```python
data.groupby("YEAR BUILT GROUP").mean().plot.bar(y="SALE\nPRICE", xlabel="Time when wasbuilt", ylabel="Sale price")
```
